## 最初に持っておきたい見方

このコースでは、式や実装を順番に覚えることよりも、『モデルは何を入力にして、何を出したいのか』を追うことが大切です。たとえば、画像を分類したいのか、次の単語を予測したいのか、行動を決めたいのかで、同じディープラーニングでも設計の中心が変わります。

## 前提知識が足りなくても進めてよい

最初から線形代数、微分、確率、PyTorch を完璧に理解している必要はありません。分からない言葉が出ても、その場で『この節は何を当てたいのか』『入力と出力は何か』『評価は何でしているか』の3点だけ追えれば、かなりの部分は読み進められます。

## このコースの読み方

1. まずは各日の intro や tutorial の最初で、問題設定を言葉で説明できるようにする
2. 次に、コードが『どの入力からどの出力を作っているか』を見る
3. 最後に、loss や accuracy などの評価値が『何の良し悪し』を表しているかを確認する

式や API 名だけを追うより、この順で読むほうが初学者には流れを掴みやすくなります。

## つまずきやすいポイント

- モデル名を覚えようとして、何の問題を解いているかを見失う
- コードの形だけを真似して、入力と出力の意味を追わない
- 精度の数字だけ見て、なぜその数字になるのかを考えない
- 学習が進んだことと、モデルが本当に理解できたことを同じだと思う

迷ったら、いま読んでいる notebook が『分類』『予測』『生成』『制御』のどれに近いのかを言い直してみると整理しやすくなります。

## 先に覚えたい最小語彙

- `input`: モデルに渡す情報
- `target`: 当てたい答え
- `model`: 入力から答えを作る関数
- `loss`: どれだけ外しているかの物差し
- `optimization`: loss を下げるようにモデルを直す手順

この5語が分かれていれば、どの tutorial でも『いま何をしているか』を追いやすくなります。

## 読み終えたときの目標

この導入を読み終えた時点では、すべての手法を知っている必要はありません。目標は、『ディープラーニングの notebook は、問題設定、入力、出力、評価の4点を押さえながら読む』という姿勢を持つことです。その姿勢があるだけで、後続の tutorial の理解はかなり安定します。
