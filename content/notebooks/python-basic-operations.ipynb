大量のデータに対して同じ操作を繰り返す場面が多く、まずは `値の持ち方` と `処理の分岐` を確実に押さえることが重要です。

このノートでは、変数・コレクション・条件分岐・反復・関数までを、実際に値を確認しながら進めます。

## 背景と目的

変数・制御構文・関数・例外処理を曖昧にしたままでは、後続の実装で小さなバグが連鎖します。

基礎操作を正確に扱えると、機械学習コードの読み書き速度とデバッグ効率が大きく上がります。

短い実行例で、文法とデータ変換の対応を確認します。


## 最初に解きたい疑問

1. f"{name}: {students}人, 完了率 {completion_ratio:.0%}" の `{:.0%}` は何をしていて、なぜ `0.45` が `45%` になるのか。
2. リスト・辞書・集合は、同じ `scores` から作っているのに何が違っていて、いつどれを選べばいいのか。
3. `scores[1:4]` の `4` はなぜ含まれないのか。`scores[-1]` と `scores[::2]` もどう読めばいいのか。
4. `for score in scores:` の中で `append` しているけれど、`labels` はいつ増えていくのか。`zip(scores, labels)` は何を表しているのか。
5. `try` で変換に失敗したとき、なぜ `except ValueError` だけを捕まえているのか。`oops` 以外で失敗したらどうなるのか。


## 先に押さえる言葉

- `動的型付け`: 変数の型を先に固定せず、実行時に入った値の種類で決まる仕組みです。Python では同じ変数に文字列や数値を入れられますが、そのぶん型の確認が重要になります。
- `f-string`: 文字列の中に変数や式をそのまま埋め込める書き方です。`f"...{name}..."` の形で使い、表示用の文を短く書けます。
- `スライス`: リストなどの一部分を切り出す書き方です。`start:end` の形で指定し、`end` 側は含まれない点がよく混乱します。
- `Counter`: 値ごとに何回出てきたかを数えるための道具です。リストを渡すと、各要素の個数をまとめた結果が返ります。


## 実行前の見取り図

1. `print(type(name), type(students), type(completion_ratio))` の結果で、文字列・整数・小数がどう見分けられるかを確認する。
2. `print(scores[1:4])` と `print(scores[::2])` の出力で、どこまで抜き出されているかを確認する。
3. `print(list(zip(scores, labels)))` と `print("変換失敗: ...")` の出力で、対応関係と失敗した要素を追う。


## つまずきやすい点

- `f"...{completion_ratio:.0%}"` や `:.2f` の書式指定が何を変えているのかを、1つずつ分解した説明が必要。
- `math.sqrt(sum((x - avg) ** 2 for x in scores) / len(scores))` が「平均との差の二乗平均の平方根」であることと、なぜその式になるのかの言い換えが必要。


```python
import math
from collections import Counter
```

`type` は「いま持っている値の性質」を確認する基本ツールです。Python は動的型付けなので、
コードを書いている途中でも型を観察しながら進めるとミスを減らせます。

```python
name = "Noema"
students = 32
completion_ratio = 0.45

print(type(name), type(students), type(completion_ratio))
print(f"{name}: {students}人, 完了率 {completion_ratio:.0%}")
```

リスト・辞書・集合は、用途が異なる 3 つの基本コンテナです。

- リスト: 順序を保って並べたい
- 辞書: キーで高速に参照したい
- 集合: 重複を除いて要素の存在を扱いたい

同じデータでも、目的に応じて構造を選ぶとコードが一気に読みやすくなります。

```python
scores = [72, 88, 65, 91, 88]
profile = {"name": "Aiko", "grade": 1, "club": "AI"}
unique_scores = set(scores)

print("scores:", scores)
print("profile['club']:", profile["club"])
print("unique_scores:", unique_scores)
```

インデックスとスライスは、配列の一部を抜き出すときの標準手段です。
とくに `start:end` は end を含まないので、境界のミスに注意します。

```python
print("先頭:", scores[0])
print("末尾:", scores[-1])
print("2番目から4番目手前:", scores[1:4])
print("2個おき:", scores[::2])
```

`if / elif / else` は条件分岐、`for` は反復処理です。
点数に応じてラベルを付ける処理を実装すると、分岐とループの関係が見えます。

出力では `zip(scores, labels)` を使っています。`zip` は同じ位置の要素を組にして、
`(score, label)` のタプル列を作る関数です。

```python
labels = []
for score in scores:
    if score >= 90:
        labels.append("A")
    elif score >= 75:
        labels.append("B")
    else:
        labels.append("C")

print(list(zip(scores, labels)))
```

リスト内包表記は、単純な変換を短く安全に書くための記法です。
まず通常の `for` を書いて意味を確認し、次に内包表記へ圧縮する流れが初心者には理解しやすいです。

```python
# 通常の for
squared_for = []
for x in range(1, 6):
    squared_for.append(x * x)

# 内包表記
squared_comp = [x * x for x in range(1, 6)]

print(squared_for)
print(squared_comp)
```

関数は「処理に名前をつけて再利用する」ための仕組みです。
入力を引数、出力を戻り値として明確にすると、分析コードの保守が楽になります。

```python
def summarize_scores(values):
    count = len(values)
    avg = sum(values) / count
    best = max(values)
    worst = min(values)
    return {"count": count, "avg": avg, "best": best, "worst": worst}

summary = summarize_scores(scores)
print(summary)
```

例外処理は「壊れる可能性がある場所」を明示するために使います。
数値変換の失敗を握りつぶすのではなく、`except` で理由を扱う習慣をつけると、後でデバッグしやすくなります。

```python
raw_inputs = ["12", "7", "oops", "20"]
parsed = []

for item in raw_inputs:
    try:
        parsed.append(int(item))
    except ValueError:
        print(f"変換失敗: {item}")

print("parsed:", parsed)
```

最後に、ここまでの要素を合わせた小さな分析パターンを確認します。
文字列ラベルの頻度集計や平均値の算出は、機械学習の前処理でも頻出です。

`Counter` は「値ごとの出現回数」を数えるためのクラスで、
`Counter(labels)` のように渡すと `{'A': 回数, 'B': 回数, ...}` を返します。

```python
label_counter = Counter(labels)
avg_score = sum(scores) / len(scores)
std_score = math.sqrt(sum((x - avg_score) ** 2 for x in scores) / len(scores))

print("label count:", label_counter)
print(f"average={avg_score:.2f}, std={std_score:.2f}")
```

Python 基礎で大切なのは、文法暗記よりも「どの構造で値を持つか」「分岐と反復をどう組み合わせるか」を自然に選べることです。
