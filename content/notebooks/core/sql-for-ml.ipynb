# 機械学習のためのSQL

機械学習の実務では、モデルに入る前の表をどう作るかで勝負がかなり決まります。その前工程の多くを担うのが SQL です。SQL は単なる抽出言語ではなく、リークを防ぎながら再現可能な学習テーブルを作るための道具です。

このノートでは、EC サービスの疑似データを題材に、基本的な SQL 操作から point-in-time の特徴量テーブル作成までを一続きで見ます。

## まずは「どの時点の何を予測したいか」を決める

SQL を機械学習向けに書くときは、`何を取るか` より先に `いつの時点で、何を予測したいか` を決める必要があります。そこが曖昧だと、JOIN や集計は書けても、未来情報が混ざった壊れた学習テーブルになりやすいからです。


このノートでは、まず `users`, `events`, `orders` の 3 テーブルを作り、基本的な抽出、集計、結合を確認します。そのあとで `snapshot` を基準に、特徴量窓とラベル窓を分けた point-in-time テーブルを作り、最後にわざとリーク版とも比べます。


## このノートで使う基本語彙

- JOIN: 別テーブルをキーでつなぐ
- GROUP BY: 同じキーごとに集計する
- snapshot: 予測時点
- 特徴量窓: snapshot より前の情報を集める期間
- ラベル窓: snapshot 以後の結果を観測する期間

ここで一番重要なのは、特徴量とラベルの時間境界を混ぜないことです。


## SQL の正しさは、時点境界で決まる

このノートでは、単にクエリが動くかではなく、「そのクエリが本番時点でも再現可能か」を見ます。同じ集計でも、`snapshot` をまたいだ瞬間に学習データの意味が壊れます。


## 先に構えたい失敗

もっとも危ないのは、未来のイベントや購入を特徴量へ混ぜてしまうことです。また、分割方法をその場しのぎで決めると、再現性のない実験になりやすくなります。

このノートでは、リークしない版とリークする版を並べて、その差をはっきり見ます。


## どこまでをこのノートで扱うか

ここでは、最小の SQL 例でリークと point-in-time 設計を理解することに集中します。本番 ETL 全体ではなく、学習テーブル作成の考え方が中心です。


## ここでは、最小の学習テーブル作成に集中する

ここで書く SQL は本番 ETL 全体を再現するものではなく、リークを避けた学習テーブルをどう考えるかを見るための最小例です。実務ではもっと列も条件も増えますが、時間境界の考え方はそのまま使えます。


## まずは、学習に使う土台テーブルを作る

最初にメモリ上のデータベースを作り、`users`, `events`, `orders` を定義します。行動ログ、購入履歴、属性情報が別テーブルに分かれているのは、実務でもかなり自然な構成です。


```python
import sqlite3
from datetime import datetime, timedelta
from random import Random
from textwrap import dedent
```

まずはメモリ上にデータベースを作り、`users`, `events`, `orders` の 3 テーブルを定義します。どのテーブルが何の情報を持つかを最初に明確にすると、あとで JOIN の意味が読みやすくなります。

```python
conn = sqlite3.connect(":memory:")
conn.row_factory = sqlite3.Row
cur = conn.cursor()

cur.executescript(
    dedent(
        """
        CREATE TABLE users (
            user_id INTEGER PRIMARY KEY,
            signup_date TEXT NOT NULL,
            plan TEXT NOT NULL,
            country TEXT NOT NULL
        );

        CREATE TABLE events (
            event_id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER NOT NULL,
            event_time TEXT NOT NULL,
            event_type TEXT NOT NULL,
            session_seconds INTEGER NOT NULL,
            clicks INTEGER NOT NULL,
            FOREIGN KEY (user_id) REFERENCES users(user_id)
        );

        CREATE TABLE orders (
            order_id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER NOT NULL,
            order_time TEXT NOT NULL,
            amount REAL NOT NULL,
            FOREIGN KEY (user_id) REFERENCES users(user_id)
        );
        """
    )
)
conn.commit()
```

## あとで何度も使う補助関数を置く

以降は SQL を何本も試すので、クエリ実行と表示の小さな補助関数を先に置いておきます。ここは主役ではなく、以後のクエリを読みやすくするための準備です。


```python
def q(sql: str, params=()):
    return conn.execute(dedent(sql), params).fetchall()


def show(rows, limit=8):
    rows = list(rows)
    if not rows:
        print("(no rows)")
        return
    cols = rows[0].keys()
    print(" | ".join(cols))
    print("-" * 100)
    for r in rows[:limit]:
        print(" | ".join(str(r[c]) for c in cols))
    if len(rows) > limit:
        print(f"... ({len(rows)} rows total)")
```

以降は `q(sql, params)` で実行し、`show(rows)` で結果を確認します。SQL の学習では、クエリを書いたらすぐ結果を見て、表の形が想像どおりかを確かめる癖が大切です。

次に、分析用の疑似データを `ユーザー生成 → イベント生成 → 注文生成` の順で入れます。どの情報がどのテーブルに属するかを意識すると、あとで集計の責任範囲が見えやすくなります。

```python
rng = Random(42)
start = datetime(2024, 1, 1)
end = datetime(2024, 6, 30)

plans = ["free", "pro", "team"]
plan_weights = [0.62, 0.30, 0.08]
countries = ["JP", "US", "IN", "DE", "FR"]

n_users = 320
for user_id in range(1, n_users + 1):
    signup = start + timedelta(days=rng.randint(0, 80))
    plan = rng.choices(plans, weights=plan_weights, k=1)[0]
    country = rng.choice(countries)
    cur.execute(
        "INSERT INTO users (user_id, signup_date, plan, country) VALUES (?, ?, ?, ?)",
        (user_id, signup.date().isoformat(), plan, country),
    )

conn.commit()
```

## まずは SQL の基本操作を、学習前処理の文脈で見る

ここから `WHERE`, `ORDER BY`, `GROUP BY`, `JOIN`, `CASE WHEN` を順に使います。単発の文法練習ではなく、「機械学習用の特徴を作る前段」として読むのがポイントです。


```python
for row in q("SELECT user_id, plan, signup_date FROM users"):
    user_id = row["user_id"]
    plan = row["plan"]
    signup_dt = datetime.fromisoformat(row["signup_date"])

    base_events = 20 if plan == "free" else 34 if plan == "pro" else 46
    n_events = max(8, int(rng.gauss(base_events, 6)))

    for _ in range(n_events):
        day_offset = rng.randint(0, 178)
        event_dt = signup_dt + timedelta(days=day_offset, hours=rng.randint(0, 23), minutes=rng.randint(0, 59))
        if event_dt > end:
            continue

        if plan == "free":
            clicks = max(0, int(rng.gauss(2.2, 1.4)))
            session_seconds = max(20, int(rng.gauss(130, 70)))
        elif plan == "pro":
            clicks = max(0, int(rng.gauss(4.8, 2.0)))
            session_seconds = max(30, int(rng.gauss(220, 90)))
        else:
            clicks = max(0, int(rng.gauss(6.5, 2.4)))
            session_seconds = max(40, int(rng.gauss(290, 110)))

        event_type = rng.choices(
            ["page_view", "search", "add_to_cart"],
            weights=[0.62, 0.25, 0.13],
            k=1,
        )[0]

        cur.execute(
            """
            INSERT INTO events (user_id, event_time, event_type, session_seconds, clicks)
            VALUES (?, ?, ?, ?, ?)
            """,
            (user_id, event_dt.isoformat(sep=" "), event_type, session_seconds, clicks),
        )

conn.commit()
```

個々のクエリでは、結果の件数や列名だけでなく「この集計は snapshot の前なのか後なのか」を毎回意識してください。SQL では、その時間境界がいちばん重要です。


```python
activity = q(
    """
    SELECT
        u.user_id,
        u.plan,
        u.signup_date,
        COALESCE(COUNT(e.event_id), 0) AS event_count,
        COALESCE(SUM(e.clicks), 0) AS total_clicks
    FROM users u
    LEFT JOIN events e ON u.user_id = e.user_id
    GROUP BY u.user_id, u.plan, u.signup_date
    """
)

for r in activity:
    user_id = r["user_id"]
    plan = r["plan"]
    signup_dt = datetime.fromisoformat(r["signup_date"])
    total_clicks = r["total_clicks"]
    event_count = r["event_count"]

    base_prob = 0.02
    if plan == "pro":
        base_prob += 0.08
    if plan == "team":
        base_prob += 0.12
    base_prob += min(0.35, total_clicks / 380.0)
    base_prob += min(0.20, event_count / 420.0)

    n_orders = 0
    for _ in range(3):
        if rng.random() < base_prob:
            n_orders += 1

    min_order_dt = signup_dt + timedelta(days=1)
    max_order_dt = min(end, signup_dt + timedelta(days=180))
    if min_order_dt > max_order_dt:
        continue

    total_seconds = int((max_order_dt - min_order_dt).total_seconds())

    for _ in range(n_orders):
        offset_sec = rng.randint(0, total_seconds)
        order_dt = min_order_dt + timedelta(seconds=offset_sec)

        mean_amount = 38 if plan == "free" else 74 if plan == "pro" else 130
        amount = max(8, round(rng.gauss(mean_amount, mean_amount * 0.35), 2))

        cur.execute(
            "INSERT INTO orders (user_id, order_time, amount) VALUES (?, ?, ?)",
            (user_id, order_dt.isoformat(sep=" "), amount),
        )

conn.commit()
```

特徴量候補を作るクエリは、見た目にはただの集計でも、機械学習の観点では「予測時点で利用可能かどうか」が本質です。そこを外すと、どれだけきれいな SQL でも意味がありません。


```python
table_counts = q(
    """
    SELECT 'users' AS table_name, COUNT(*) AS n FROM users
    UNION ALL
    SELECT 'events' AS table_name, COUNT(*) AS n FROM events
    UNION ALL
    SELECT 'orders' AS table_name, COUNT(*) AS n FROM orders
    """
)
show(table_counts)
```

ここから SQL の基本操作を確認します。まずは `WHERE` と `ORDER BY` で、特定ユーザーのイベント履歴を時間順に取り出します。行動ログを「いつ何をしたか」で読む感覚をここで掴みます。

```python
rows = q(
    """
    SELECT user_id, event_time, event_type, session_seconds, clicks
    FROM events
    WHERE user_id = ?
    ORDER BY event_time
    LIMIT 10
    """,
    (12,),
)
show(rows)
```

次に `GROUP BY` でユーザー単位の集計を作ります。イベントをそのまま見るだけではなく、回数や合計へ要約すると、学習モデルに渡しやすい形へ近づきます。

```python
agg = q(
    """
    SELECT
        user_id,
        COUNT(*) AS n_events,
        SUM(clicks) AS total_clicks,
        AVG(session_seconds) AS avg_session_seconds
    FROM events
    GROUP BY user_id
    HAVING COUNT(*) >= 18
    ORDER BY total_clicks DESC
    LIMIT 10
    """
)
show(agg)
```

`JOIN` と `CASE WHEN` を使うと、属性と行動を混ぜた特徴量を作れます。ここでは、ある時点より前 90 日間の購入金額とプラン情報を合わせて、学習に使える 1 行へ寄せていきます。

```python
asof_ts = '2024-05-01 00:00:00'

joined = q(
    """
    WITH spend_90 AS (
      SELECT
          user_id,
          SUM(amount) AS spend_90d
      FROM orders
      WHERE order_time >= datetime(:asof_ts, '-90 day')
        AND order_time < :asof_ts
      GROUP BY user_id
    )
    SELECT
        u.user_id,
        u.plan,
        CASE WHEN u.plan = 'free' THEN 0 ELSE 1 END AS paid_flag,
        COALESCE(s.spend_90d, 0) AS spend_90d
    FROM users u
    LEFT JOIN spend_90 s ON u.user_id = s.user_id
    ORDER BY spend_90d DESC
    LIMIT 12
    """,
    {"asof_ts": asof_ts},
)
show(joined)
```

ここから機械学習用テーブルを作ります。`snapshot` を予測時点とし、その前 30 日を特徴量窓、その後 30 日をラベル窓とします。

この境界を明示した瞬間に、SQL は単なる集計ではなく point-in-time の学習テーブル設計になります。


```python
snapshot = '2024-05-01 00:00:00'
horizon_days = 30

feature_sql = """
WITH event_30d AS (
    SELECT
        user_id,
        COUNT(*) AS events_30d,
        SUM(clicks) AS clicks_30d,
        AVG(session_seconds) AS avg_session_30d,
        MAX(event_time) AS last_event_time
    FROM events
    WHERE event_time < :snapshot
      AND event_time >= datetime(:snapshot, '-30 day')
    GROUP BY user_id
),
order_90d AS (
    SELECT
        user_id,
        SUM(amount) AS spend_90d
    FROM orders
    WHERE order_time < :snapshot
      AND order_time >= datetime(:snapshot, '-90 day')
    GROUP BY user_id
),
label_window AS (
    SELECT
        user_id,
        CASE WHEN COUNT(*) > 0 THEN 1 ELSE 0 END AS purchased_30d
    FROM orders
    WHERE order_time >= :snapshot
      AND order_time < datetime(:snapshot, '+' || :horizon_days || ' day')
    GROUP BY user_id
)
SELECT
    u.user_id,
    u.country,
    CASE WHEN u.plan = 'free' THEN 0 ELSE 1 END AS paid_flag,
    COALESCE(e.events_30d, 0) AS events_30d,
    COALESCE(e.clicks_30d, 0) AS clicks_30d,
    COALESCE(e.avg_session_30d, 0) AS avg_session_30d,
    -- イベントがないユーザーは signup_date を最終行動日として扱う
    CAST((julianday(:snapshot) - julianday(COALESCE(e.last_event_time, u.signup_date))) AS INTEGER) AS days_since_last_event,
    COALESCE(o.spend_90d, 0) AS spend_90d,
    COALESCE(l.purchased_30d, 0) AS label
FROM users u
LEFT JOIN event_30d e ON u.user_id = e.user_id
LEFT JOIN order_90d o ON u.user_id = o.user_id
LEFT JOIN label_window l ON u.user_id = l.user_id
ORDER BY u.user_id
"""
```

## リークしない版と、リークする版を比べる

ここがこのノートの核心です。未来を含まない `clean` な特徴量表と、未来を少しだけ見てしまう `leaky` な特徴量表を並べると、SQL の条件 1 つで学習の意味がどう壊れるかが分かります。


```python
feature_rows = q(feature_sql, {"snapshot": snapshot, "horizon_days": horizon_days})
show(feature_rows, limit=12)
print('rows:', len(feature_rows))
```

境界条件の違いはわずかでも、ラベル窓の情報が特徴量へ混ざるとモデルは先回りで当てられてしまいます。リークは SQL の書き間違いとして起こるので、モデリング以前の問題です。


```python
label_dist_sql = f"""
WITH base AS (
{feature_sql}
)
SELECT
    label,
    COUNT(*) AS n_users,
    ROUND(AVG(events_30d), 2) AS avg_events_30d,
    ROUND(AVG(spend_90d), 2) AS avg_spend_90d
FROM base
GROUP BY label
ORDER BY label
"""

label_dist = q(label_dist_sql, {"snapshot": snapshot, "horizon_days": horizon_days})
show(label_dist)
```

次に、リークの例を確認します。比較を明確にするため、窓幅は同じ 30 日に固定し、未来を含むかどうかだけを変えます。こうすると、差の原因が時間境界にあることがはっきり見えます。

```python
clean_stats_sql = f"""
WITH base AS (
{feature_sql}
)
SELECT label, ROUND(AVG(events_30d), 2) AS avg_events
FROM base
GROUP BY label
ORDER BY label
"""

leaky_stats_sql = """
WITH event_30d_leaky AS (
    SELECT
        user_id,
        COUNT(*) AS events_30d
    FROM events
    WHERE event_time >= :snapshot
      -- NG: snapshot以降（未来）のイベントを特徴量に含める
      AND event_time < datetime(:snapshot, '+' || :horizon_days || ' day')
    GROUP BY user_id
),
label_window AS (
    SELECT
        user_id,
        CASE WHEN COUNT(*) > 0 THEN 1 ELSE 0 END AS label
    FROM orders
    WHERE order_time >= :snapshot
      AND order_time < datetime(:snapshot, '+' || :horizon_days || ' day')
    GROUP BY user_id
)
SELECT
    COALESCE(l.label, 0) AS label,
    ROUND(AVG(COALESCE(e.events_30d, 0)), 2) AS avg_events
FROM users u
LEFT JOIN event_30d_leaky e ON u.user_id = e.user_id
LEFT JOIN label_window l ON u.user_id = l.user_id
GROUP BY COALESCE(l.label, 0)
ORDER BY label
"""

clean_stats = q(clean_stats_sql, {"snapshot": snapshot, "horizon_days": horizon_days})
leaky_stats = q(leaky_stats_sql, {"snapshot": snapshot, "horizon_days": horizon_days})

print('clean feature mean by label')
show(clean_stats)
print('leaky feature mean by label')
show(leaky_stats)
```

`clean` と `leaky` の差は、未来窓のイベントを見た瞬間に「まだ起きていない購買」を知ってしまうことを示しています。SQL の条件は些細に見えても、snapshot 境界をまたいだ時点で学習データ全体が壊れます。


最後に、学習用テーブルを分割できる形へ整えます。ここでは説明を単純にするため `user_id % 5` で fold を作ります。実務ではハッシュを使うことが多いですが、「再現可能な分割を SQL 側で持てる」ことがポイントです。

```python
dataset_fold_sql = f"""
WITH base AS (
{feature_sql}
)
SELECT
    user_id,
    paid_flag,
    events_30d,
    clicks_30d,
    avg_session_30d,
    days_since_last_event,
    spend_90d,
    label,
    (user_id % 5) AS fold_id
FROM base
ORDER BY user_id
"""

dataset_with_fold = q(dataset_fold_sql, {"snapshot": snapshot, "horizon_days": horizon_days})
show(dataset_with_fold, limit=12)
```

SQL は機械学習パイプラインの前処理ではなく、学習品質を決める中核です。とくに重要なのは、ラベル時点と特徴量時点の境界を明示し、未来情報を混ぜないことです。モデルを変える前に、SQL で作る学習テーブルが本番でも再現できるかを必ず疑ってください。
