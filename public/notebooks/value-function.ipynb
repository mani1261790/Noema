# 価値関数

価値関数は「この状態にいると将来どれだけ得か」を数値で表す概念です。方策の良し悪しを比較する基準になるため、強化学習の中心に置かれます。

## 背景と目的

行動を直接比較する前に、状態自体の期待リターンを見積もると方策評価が安定します。

価値関数を理解できると、ベルマン方程式・方策反復・価値反復の流れを一貫して読めるようになります。

ここでは有限長エピソードとマルコフ連鎖の2通りで状態価値を計算します。


```python
import numpy as np

gamma = 0.9

# 2状態 + 終端の方策固定マルコフ連鎖
states = ['s0', 's1']
P_pi = np.array([
    [0.6, 0.4],
    [0.3, 0.7],
])
R_pi = np.array([0.2, 0.8])
```

## 解析解で価値を求める

固定方策 \(\pi\) の状態価値は、

\[V^{\pi}=R_{\pi}+\gamma P_{\pi}V^{\pi}\]

を満たします。したがって

\[V^{\pi}=(I-\gamma P_{\pi})^{-1}R_{\pi}\]

で求まります。


```python
I = np.eye(len(states))
V_exact = np.linalg.solve(I - gamma * P_pi, R_pi)
print('V^pi (exact) =', {s: round(v, 6) for s, v in zip(states, V_exact)})
```

## 反復評価で同じ値に近づくことを確認

解析解だけでなく、ベルマン期待更新を反復して同じ値に収束することを確かめます。


```python
V = np.zeros(len(states))
for it in range(40):
    V = R_pi + gamma * (P_pi @ V)
print('V^pi (iterative) =', {s: round(v, 6) for s, v in zip(states, V)})
print('max abs diff      =', round(float(np.max(np.abs(V - V_exact))), 8))
```

状態価値を安定して推定できると、次段階で行動価値へ拡張したときも更新の意味を保ったまま実装できます。
