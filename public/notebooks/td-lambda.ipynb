# TD(λ)

TD(λ) は 1-step TD とモンテカルロの間を連続的につなぐ方法です。`λ` を動かすことで、短期更新と長期更新の混合比を滑らかに調整できます。

## 背景と目的

`λ=0` は 1-step TD、`λ→1` は長い将来報酬を強く参照する更新になります。

タスクに応じて `λ` を調整できると、学習速度と安定性の両立がしやすくなります。

前向き視点の λ-return と後ろ向き視点の trace 更新を同じ軌跡で比較します。


## 最初に解きたい疑問

1. `lambda_return` は、`n-step return` をどう混ぜているのか。
2. `λ=0` と `λ=0.95` の差を、どこで見ればよいのか。
3. eligibility trace の `E` は、何を覚えているのか。
4. 前向き視点と後ろ向き視点は、どう対応しているのか。
5. `Final trace` を見て、何を理解すればよいのか。


## 先に押さえる言葉

- `lambda-return`: 複数の n-step return を重み付きで混ぜた目標値です。`λ` で短期と長期の比率を調整します。
- `eligibility trace`: どの過去状態をどれだけ更新対象にするかを表す履歴です。TD誤差を過去へ配る役目を持ちます。
- `forward view`: 将来の報酬列を見て、理想的な目標値を作る見方です。`lambda_return` がこれに近いです。
- `backward view`: 現在のTD誤差を、過去の状態に trace で配る見方です。オンライン実装でよく使います。

- `λ→1 の極限`: 終端までの full return を使う設定では、Monte Carlo return に近づきます。終端で bootstrap を切ることが前提です。


## 実行前の見取り図

1. `lambda=` ごとの `G^lambda` が変化しているか。
2. `Updated V = ...` で、trace を使った更新結果が出ているか。
3. `Final trace = ...` で、過去状態の残り方が見えるか。


## つまずきやすい点

- 重み `w` がなぜその形になるのかの説明がない。
- `lambda-return` と trace 更新の同値性を、途中式つきでつなぐ説明が不足している。


```python
import math

gamma = 0.9
rewards = [0.4, -0.2, 0.8, 1.1]
v_boot = 0.5


def n_step_return(rews, n, gamma, v_boot):
    n = min(n, len(rews))
    g = 0.0
    for k in range(n):
        g += (gamma ** k) * rews[k]
    if n < len(rews):
        g += (gamma ** n) * v_boot
    return g


def lambda_return(rews, gamma, lam, v_boot):
    T = len(rews)
    out = 0.0
    for n in range(1, T + 1):
        w = (1 - lam) * (lam ** (n - 1)) if n < T else (lam ** (n - 1))
        out += w * n_step_return(rews, n, gamma, v_boot)
    return out

for lam in [0.0, 0.3, 0.7, 0.95]:
    print('lambda=', lam, 'G^lambda=', round(lambda_return(rewards, gamma, lam, v_boot), 6))
```

## 後ろ向き視点（Eligibility Trace）

\[e_t(s)=\gamma\lambda e_{t-1}(s)+\mathbf{1}[s_t=s],\quad V(s)\leftarrow V(s)+\alpha\,\delta_t\,e_t(s)\]

前向き計算を毎回やり直さず、オンラインで近い効果を得る実装です。


```python
alpha = 0.15
lam = 0.7
states = ['s0', 's1', 's0', 's2']
rewards = [0.3, 0.1, 0.7, 0.0]
V = {'s0': 0.2, 's1': 0.4, 's2': 0.1}
E = {'s0': 0.0, 's1': 0.0, 's2': 0.0}

for t in range(len(states) - 1):
    s = states[t]
    s_next = states[t + 1]
    r = rewards[t]
    delta = r + gamma * V[s_next] - V[s]

    for k in E:
        E[k] *= gamma * lam
    E[s] += 1.0

    for k in V:
        V[k] += alpha * delta * E[k]

print('Updated V =', {k: round(v, 6) for k, v in V.items()})
print('Final trace =', {k: round(v, 6) for k, v in E.items()})
```
